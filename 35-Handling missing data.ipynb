### cca(Complete case analysis) is applied if only 5% data are missing at random 

```python
import pandas as pd 
import numpy as np 
```

```python
df=pd.read_csv(r"https://raw.githubusercontent.com/campusx-official/100-days-of-machine-learning/main/day35-complete-case-analysis/data_science_job.csv")
```

```python
df.head()
```

###  find out the cols less than 4 or 5% data is null and remove them 

```python
col=[var for var in df.columns if df[var].isnull().mean()<0.05 and df[var].isnull().mean()>0]
```

```python
col
```

```python
df[col].dropna()
```
